```python
import pandas as pd
from datetime import date,datetime, timedelta
import numpy as np
```

```python
nacional = pd.read_csv("./884e4edb6c3f1cda0fbc930dbed29050_Download_COVID19_20200412.csv", sep=';')
print(nacional.shape)
nacional.head()
```

```python
roraima = nacional.query("estado == 'RR'")
roraima
```

```python
roraima['data'] = pd.to_datetime(roraima['data'], format='%d/%m/%Y')
roraima = roraima.sort_values(by='data', axis = 0)
date_str = []
for dt in roraima['data']:
    conv = datetime.date(dt)
    date_str.append(conv.strftime("+%Y-%m-%dT00:00:00Z/11"))
roraima['wdt_dates'] = date_str
roraima
```

```python
roraima_cases = roraima.query("casosAcumulados != 0")
roraima_cases
```
